```python
def date_to_nice_string(my_date):
    my_date = datetime.strptime(my_date, "%Y-%m-%dT%H:%M:%S.%fZ")
    month = [
        "Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
    ]
    return f"{month[my_date.month - 1]} {my_date.day}, {my_date.year}"

def date_to_unix(date_string):
    return datetime.strptime(date_string, "%Y-%m-%dT%H:%M:%S.%fZ").timestamp()
```

```python
def get_blogpost_content(postid):
    dynamodb = boto3.client('dynamodb', region_name=AWS_REGION, aws_access_key_id=AWS_ACCESS_KEY, aws_secret_access_key=AWS_SECRET_KEY)

    params = {
        'TableName': AWS_DYNAMO_TABLE,
        'Key': {
            'post_folder': {'S': postid}
        }
    }

    blogpost_content = dynamodb.get_item(**params).get("Item", [])
    
    return blogpost_content
```

```python
s3_client = boto3.client('s3', aws_access_key_id=AWS_ACCESS_KEY, aws_secret_access_key=AWS_SECRET_KEY, region_name=AWS_REGION)

def upload_to_s3(file_path, s3_key, acl='public-read'):
    s3_client.upload_file(file_path, BUCKET_NAME, s3_key, ExtraArgs = {'ContentType': 'text/html; charset=utf-8'})

def save_and_upload_html(html_content, local_path, s3_key):
    # Save the HTML content to a file
    with open(local_path, "w", encoding="utf-8") as file:
        file.write(html_content)

    # Upload the file to S3
    upload_to_s3(local_path, s3_key)
```

```python
from datetime import datetime
from io import BytesIO
def create_sitemap():
    event = datetime.utcnow()
    dynamodb = boto3.client('dynamodb', region_name=AWS_REGION, aws_access_key_id=AWS_ACCESS_KEY, aws_secret_access_key=AWS_SECRET_KEY)
    
    # Assuming AWS_DYNAMO_TABLE, SITE_CANONICAL_URL, BLOG_SEO_SUBFOLDER, AWS_S3_BUCKET are defined elsewhere
    params = {
        "TableName": AWS_DYNAMO_TABLE,
        "ProjectionExpression": "title, post_folder, post_body, date_modified",
    }

    # Scan all blog posts data
    response = dynamodb.scan(**params)
    lastmod_value = event.strftime('%Y-%m-%dT%H:%M:%S.%f')[:-3] + 'Z'
    posts = f'''<?xml version="1.0" encoding="UTF-8"?>
                <urlset xmlns="http://www.sitemaps.org/schemas/sitemap/0.9">
                    <url>
                        <loc>{SITE_CANONICAL_URL}</loc>
                        <lastmod>{lastmod_value}</lastmod>
                        <changefreq>monthly</changefreq>
                        <priority>0.8</priority>
                    </url>
                '''

    for blogpost in response.get("Items", []):
        posts += f'''<url>
                        <loc>{SITE_CANONICAL_URL}{BLOG_SEO_SUBFOLDER}/{blogpost["post_folder"]["S"]}/index.html</loc>
                        <lastmod>{blogpost["date_modified"]["S"]}</lastmod>
                        <changefreq>monthly</changefreq>
                    </url>'''

    posts += "</urlset>"

    s3_client = boto3.client("s3")

    # Upload the sitemap to S3
    s3_client.upload_fileobj(
        Fileobj=BytesIO(posts.encode('utf-8')),
        Bucket=BUCKET_NAME,
        Key="sitemap.xml",
        ExtraArgs={"ContentType": "application/xml; charset=utf-8"},
    )

    print("Successfully uploaded your sitemap.")
```
